# Build Android APK
1. Run the cell below.
2. Enter your GitHub Repository URL when prompted (e.g., https://github.com/Username/RepoName).
3. The APK will download automatically upon completion.

```python
import os
from google.colab import files

# 1. Input Repository URL
repo_url = input("Enter GitHub Repository URL: ").strip()

if repo_url:
    repo_name = repo_url.split("/")[-1].replace(".git", "")
    
    # 2. Clone Repo
    !git clone {repo_url}
    
    if os.path.exists(repo_name):
        os.chdir(repo_name)
        
        # 3. Build APK
        print("Building APK... Please wait.")
        !chmod +x gradlew
        !./gradlew assembleDebug
        
        # 4. Download APK
        apk_path = "app/build/outputs/apk/debug/app-debug.apk"
        if os.path.exists(apk_path):
            print("Build Successful! Downloading APK...")
            files.download(apk_path)
        else:
            print("Build Failed. Check logs above.")
    else:
        print("Failed to clone repository.")
else:
    print("No URL provided.")
```
